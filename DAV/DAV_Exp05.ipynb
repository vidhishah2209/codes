```python
import pandas as pd

df = pd.read_csv("students_performance.csv")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.columns)
print(df.shape)
```

```python
print(df.isnull().sum())
df.dropna()
df['Math_Score'].fillna(df['Math_Score'].mean(), inplace=True)
df['Science_Score'].fillna(df['Science_Score'].median(), inplace=True)
df['English_Score'].fillna(df['English_Score'].mode()[0], inplace=True)
df[df.duplicated()]
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True, bins=5, color='skyblue')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Study_Hours', y='Math_Score', hue='Age', palette='Set2')
plt.title('Study Hours vs. Math Score')
plt.xlabel('Study Hours')
plt.ylabel('Math Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Age', y='Science_Score', palette='viridis')
plt.title('Average Science Score by Age')
plt.xlabel('Age')
plt.ylabel('Average Science Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Attendance', y='English_Score', hue='Age', palette='coolwarm')
plt.title('Attendance vs. English Score')
plt.xlabel('Attendance (%)')
plt.ylabel('English Score')
plt.show()
```
