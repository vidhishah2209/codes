```python
import requests
import pandas as pd
import json
```

```python
url = "https://catfact.ninja/fact"
response = requests.get(url)
data = response.json()
```

```python
print("RAW JSON DATA:\n")
print(json.dumps(data, indent=2))
```

```python
cat_fact = data['fact']
cat_length = data['length']
```

```python
df = pd.DataFrame({
    "fact": [cat_fact],
    "length": [cat_length]
})
```

```python
df.to_csv("cat_fact.csv", index=False)
```
