```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
img_bgr = cv2.imread('dog.jpeg')
plt.imshow(img_bgr)
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
```

5. POWER LAW TRANSFORMATION

```python
gamma = 0.2
r = img_gray

s = np.power(r, gamma)
plt.imshow(s, cmap='gray')
```

```python
gamma = 1
r = img_gray

s = np.power(r, gamma)
plt.imshow(s, cmap='gray')
```

```python
gamma = 1.5
r = img_gray

s = np.power(r, gamma)
plt.imshow(s, cmap='gray')
```

6. DYNAMIC RANGE COMPRESSION(LOG TRANSFORMATION)

```python
c = 0.5
r = img_gray

s = c * np.log(1 + abs(r))
plt.imshow(s, cmap='gray')
```

```python
c = 14000
r = img_gray

s = c * np.log(1 + abs(r))
plt.imshow(s, cmap='gray')
```
