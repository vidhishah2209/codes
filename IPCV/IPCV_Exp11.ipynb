```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("hpf1.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap="gray")
```

```python
ksize1 = (3, 3)
sigma1 = 1
sigma2 = 5
g1 = cv2.GaussianBlur(img, ksize1, sigma1)
plt.imshow(g1, cmap="gray")
```

```python
g2 = cv2.GaussianBlur(img, ksize1, sigma2)
plt.imshow(g2, cmap="gray")
```

```python
dog = g1 - g2
plt.imshow(dog, cmap="gray")
```

```python
ksize1 = (3, 3)
ksize2 = (5, 5)
sigma2 = 5.0
g1 = cv2.GaussianBlur(img, ksize1, sigma1)
g2 = cv2.GaussianBlur(img, ksize2, sigma1)
dog = g1 - g2
plt.imshow(g1, cmap="gray")
```
