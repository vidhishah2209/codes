```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_bgr = cv2.imread('horizontal.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img, cmap='gray')
```

HORIZONTAL MASK

```python
horizontal_mask = np.array([[-1, -1, -1], [2, 2, 2], [-1, -1, -1]], dtype=np.float32)
horizontal_img = np.zeros(img.shape, dtype=np.float32)

for i in range(1, img.shape[0]-1):
    for j in range(1, img.shape[1]-1):
        horizontal_img[i,j] = img[i-1,j-1]*horizontal_mask[0][0] + img[i-1,j]*horizontal_mask[0][1] + img[i-1,j+1]*horizontal_mask[0][2] + img[i,j-1]*horizontal_mask[1][0] + img[i,j]*horizontal_mask[1][1] + img[i,j+1]*horizontal_mask[1][2] + img[i+1,j-1]*horizontal_mask[2][0] + img[i+1,j]*horizontal_mask[2][1] + img[i+1,j+1]*horizontal_mask[2][2]

horizontal_img = np.clip(horizontal_img, 0, 255).astype(np.uint8)

plt.imshow(horizontal_img, cmap='gray')
```

VERTICAL MASK

```python
vertical_mask = np.array([[-1, 2, -1], [-1, 2, -1], [-1, 2, -1]], dtype=np.float32)
vertical_img = np.zeros(img.shape, dtype=np.float32)

for i in range(1, img.shape[0]-1):
    for j in range(1, img.shape[1]-1):
        vertical_img[i,j] = img[i-1,j-1]*vertical_mask[0][0] + img[i-1,j]*vertical_mask[0][1] + img[i-1,j+1]*vertical_mask[0][2] + img[i,j-1]*vertical_mask[1][0] + img[i,j]*vertical_mask[1][1] + img[i,j+1]*vertical_mask[1][2] + img[i+1,j-1]*vertical_mask[2][0] + img[i+1,j]*vertical_mask[2][1] + img[i+1,j+1]*vertical_mask[2][2]

vertical_img = np.clip(vertical_img, 0, 255).astype(np.uint8)

plt.imshow(vertical_img, cmap='gray')
```
