```python
img = cv2.imread("binary.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
img.shape
```

DILATION

```python
kernel = np.ones((5, 5), np.uint8)
img_dilation = cv2.dilate(img, kernel, iterations=1)
plt.imshow(img_dilation, cmap='gray')
```

EROSION

```python
img_erosion = cv2.erode(img, kernel, iterations=1)
plt.imshow(img_erosion, cmap='gray')
```

OPENING

```python
img_open = cv2.dilate(img_erosion, kernel, iterations=1)
plt.imshow(img_open, cmap='gray')
```

CLOSING

```python
img_close = cv2.erode(img_dilation, kernel, iterations=1)
plt.imshow(img_close, cmap='gray')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

axes[0, 0].imshow(img_dilation, cmap='gray')
axes[0, 0].set_title('Dilation')
axes[0, 0].axis('off')

axes[0, 1].imshow(img_erosion, cmap='gray')
axes[0, 1].set_title('Erosion')
axes[0, 1].axis('off')

axes[1, 0].imshow(img_open, cmap='gray')
axes[1, 0].set_title('Opening')
axes[1, 0].axis('off')

axes[1, 1].imshow(img_close, cmap='gray')
axes[1, 1].set_title('Closing')
axes[1, 1].axis('off')

# Adjust layout to prevent titles from overlapping
plt.tight_layout()

# Display the figure with the images
plt.show()
```
