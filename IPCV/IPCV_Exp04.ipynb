```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
img = cv2.imread('hpf1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img, cmap='gray')
```

```python
img
```

HIGH PASS FILTER

```python
hpf_mask = np.array([[-1, -1, -1],
                      [-1, 8, -1],
                      [-1, -1, -1]])
hpf_mask = hpf_mask/9
print(hpf_mask)
```

```python
hpf_img = np.zeros(img.shape, dtype="uint8")
for i in range(1, img.shape[0]-2):
    for j in range(1, img.shape[1]-2):
        hpf_img[i, j] = np.sum(np.multiply(img[i-1:i+2, j-1:j+2], hpf_mask))
```

```python
plt.imshow(hpf_img, cmap='gray')
```

HIGH BOOST FILTER

```python
hbf_mask = np.array([[-1, -1, -1],
                      [-1, 9.8, -1],
                      [-1, -1, -1]])
hbf_mask = hbf_mask/9
print(hbf_mask)
```

```python
hbf_img = np.zeros(img.shape, dtype="uint8")
for i in range(1, img.shape[0]-2):
  for j in range(1, img.shape[1]-2):
    hbf_img[i, j] = np.sum(np.multiply(img[i-1:i+2, j-1:j+2], hbf_mask))
```

```python
plt.imshow(hbf_img, cmap='gray')
```
