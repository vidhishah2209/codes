```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_bgr = cv2.imread('/content/diagonal.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img, cmap='gray')
```

PREWITT MASK

```python
fx = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]], dtype=np.float32)
r1 = np.zeros(img.shape, dtype=np.float32)
for i in range(1, img.shape[0]-1):
    for j in range(1, img.shape[1]-1):
      r1[i, j] = img[i-1,j-1]*fx[0][0] + img[i-1,j]*fx[0][1] + img[i-1,j+1]*fx[0][2] + img[i,j-1]*fx[1][0] + img[i,j]*fx[1][1] + img[i,j+1]*fx[1][2] + img[i+1,j-1]*fx[2][0] + img[i+1,j]*fx[2][1] + img[i+1,j+1]*fx[2][2]

plt.imshow(r1, cmap='gray')
```

```python
fy = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=np.float32)
r2 = np.zeros(img.shape, dtype=np.float32)
for i in range(1, img.shape[0]-1):
    for j in range(1, img.shape[1]-1):
      r2[i, j] = img[i-1,j-1]*fy[0][0] + img[i-1,j]*fy[0][1] + img[i-1,j+1]*fy[0][2] + img[i,j-1]*fy[1][0] + img[i,j]*fy[1][1] + img[i,j+1]*fy[1][2] + img[i+1,j-1]*fy[2][0] + img[i+1,j]*fy[2][1] + img[i+1,j+1]*fy[2][2]

plt.imshow(r2, cmap='gray')
```

```python
r3 = np.zeros(img.shape, dtype=np.float32)
r3 = r1 + r2
```

```python
plt.imshow(r3, cmap='gray')
```

```python
f = fx + fy
r4 = np.zeros(img.shape, dtype=np.float32)
for i in range(1, img.shape[0]-1):
    for j in range(1, img.shape[1]-1):
      r4[i, j] = img[i-1,j-1]*f[0][0] + img[i-1,j]*f[0][1] + img[i-1,j+1]*f[0][2] + img[i,j-1]*f[1][0] + img[i,j]*f[1][1] + img[i,j+1]*f[1][2] + img[i+1,j-1]*f[2][0] + img[i+1,j]*f[2][1] + img[i+1,j+1]*f[2][2]

plt.imshow(r4, cmap='gray')
```
