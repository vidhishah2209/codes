```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_bgr = cv2.imread("dog.jpeg")
plt.imshow(img_bgr)
```

```python
img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img1, cmap="gray")
```

```python
img1 = cv2.resize(img1, (350, 300))
plt.imshow(img1, cmap="gray")
```

1. DIGITAL NEGATIVE

```python
L = 256
r = img1
s = (L-1) - r
plt.imshow(s, cmap="gray")
```

2. THRESHOLDING

```python
a1 = 50
rows = img1.shape[0]
columns = img1.shape[1]
r = img1

for i in range(rows):
  for j in range(columns):
    if r[i][j] <= a1:
      s[i][j] = 0
    else:
      s[i][j] = L - 1
plt.imshow(s, cmap="gray")
```

```python
a2 = 150
rows = img1.shape[0]
columns = img1.shape[1]
r = img1

for i in range(rows):
  for j in range(columns):
    if r[i][j] <= a2:
      s[i][j] = 0
    else:
      s[i][j] = L - 1
plt.imshow(s, cmap="gray")
```

3. GREY-LEVEL SLICING


i.a. Without Background

```python
a1 = 50
b1 = 100

rows = img1.shape[0]
columns = img1.shape[1]

r = img1

for i in range(rows):
  for j in range(columns):
    if a1 <= r[i][j] <= b1:
      s[i][j] = L - 1
    else:
      s[i][j] = 0
plt.imshow(s, cmap="gray")
```

```python
a2 = 70
b2 = 150

rows = img1.shape[0]
columns = img1.shape[1]

r = img1

for i in range(rows):
  for j in range(columns):
    if a1 <= r[i][j] <= b1:
      s[i][j] = 0
    else:
      s[i][j] = L - 1
plt.imshow(s, cmap="gray")
```

ii.a. With background

```python
a1 = 50
b1 = 100

rows = img1.shape[0]
columns = img1.shape[1]

r = img1

for i in range(rows):
  for j in range(columns):
    if a1 <= r[i][j] <= b1:
      s[i][j] = L - 1
    else:
      s[i][j] = r[i][j]
plt.imshow(s, cmap="gray")
```

ii.b. With background

```python
a1 = 50
b1 = 100

rows = img1.shape[0]
columns = img1.shape[1]

r = img1

for i in range(rows):
  for j in range(columns):
    if a1 <= r[i][j] <= b1:
      s[i][j] = 0
    else:
      s[i][j] = r[i][j]
plt.imshow(s, cmap="gray")
```

4. CONTRAST STRETCHING

```python
m = 2.5
l = 0.3
n = 0.6
a = 45
b = 120
v = a * l
w = v + m*(b - a)
rows = img1.shape[0]
columns = img1.shape[1]
r = img1
```

```python
for i in range(rows):
  for j in range(columns):
    if 0 <= r[i][j] <= a:
      s[i][j] = l * r[i][j]
    elif a < r[i][j] <= b:
      s[i][j] = m*(r[i][j] - a) + v
    else:
      s[i][j] = n*(r[i][j] - b) + w

plt.imshow(s, cmap="gray")
```
